### 구별 휘발유 가격에 대한 데이터 정리

```python
import pandas as pd
from glob import glob
```

```python
glob('./station_data/지역*.xls')
```

```python
df = pd.read_excel('./station_data/지역_위치별(주유소).xls', header=2)
df.head()
```

```python
station_files = glob('./station_data/지역*.xls')
tmp_raw = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
station_raw.info()
```

```python
stations = pd.DataFrame({
    '상호': station_raw['상호'],
    '주소': station_raw['주소'],
    '가격': station_raw['휘발유'],
    '셀프': station_raw['셀프여부'],
    '상표': station_raw['상표']
})
stations.head()
```

```python
stations['구'] = [addr.split()[1] for addr in stations['주소']]
stations.tail()
```

```python
stations['구'].unique()
```

```python
stations = stations[stations['가격'] != '-']
stations.info()
```

```python
stations['가격'] = pd.to_numeric(stations['가격'])
stations.info()
```

### 셀프 주유소가 정말 저렴한지 Boxplot으로 확인하기

```python
import seaborn as sns 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
stations.boxplot(column='가격', by='셀프', figsize=(12, 8))
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='상표', y='가격', hue='셀프', data=stations, palette='Set3')
plt.show()
```
