```python
age = pd.cut(titanic['age'], [1, 20, 30, 50, 70, 100])
titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
age = pd.cut(titanic['age'], [1, 20, 30, 50, 70, 100])
titanic.pivot_table('survived', ['who', age], 'class')
```

```python
fare = pd.qcut(titanic['fare'], 3)
titanic.pivot_table('survived', ['who', age], [fare, 'class'])
```
